# Learning Stack for Exercises with the new LangChain 1.0 Ecosystem

## Make sure you have Python 3.13.11 installed in your computer
* **See the “Lessons for your Second Round” to learn how to do it**.

#### For macOS Users (this the option we select, use the best one for you)
* Update Homebrew and pyenv Open your Terminal and run:

`brew update && brew upgrade pyenv`

* Check for Python 3.13.11

`pyenv install --list | grep 3.13`

* Install Python 3.13.11

`pyenv install 3.13.11`

That’s it! You can now use this Python version for your projects.

## Make sure you have Poetry 2.2.1 installed in your computer
* You might have Poetry 1.8.2 from the initial learning stack, but want to try a newer version (like 2.2.1). With pipx, you can safely install and use multiple Poetry versions at the same time!

* **See the “Lessons for your Second Round” to learn how to do it**.

### Install the new Poetry version alongside your old one, as a separate command:
`pipx install poetry==2.2.1 --suffix=-2.2.1`

* This creates a new command: `poetry-2.2.1`


#### Make sure you have both versions available
`poetry --version`         # (should show 1.8.2 or your old version)
`poetry-2.2.1 --version`  # (should show 2.2.1)

#### Use the Right Version for Each Project
For the projects that require Poetry 1.8.2, just keep using the regular `poetry` command (your old version).

For the projects of the new LangChain 1.0 Ecosystem, use `poetry-2.2.1`.

## You are now ready to work with the exercises files of the new LangChain 1.0 Ecosystem
You will follow the same process you learned in the previous lessons for the regular exercises:
* Step 1: Download or clone the Github Repository.
* Step 2: Navigate to the Project Directory.
* Step 3: Make sure you have the right Python version active in the directory.
    * `pyenv local 3.13.11` 
* Step 4: Install Dependencies with Poetry.
    * `poetry-2.2.1 install`
* Step 5: Activate the Virtual Environment with Poetry (aka the Poetry Shell).
    * Remember that starting with Poetry 2.x, the way you activate your project’s virtual environment has changed. Now we have to do:
    * `poetry-2.2.1 env activate`
    * When you run this command inside your project directory, Poetry will display the command you should use to activate the environment in your current shell. For example, it may output something like:
        * `source /Users/youruser/Library/Caches/pypoetry/virtualenvs/amazingproject-XXXX-py3.13/bin/activate`
    * Copy and run that command in your terminal.
    * Now, your terminal session is inside the project’s virtual environment: any Python commands you run, or any packages you install (for example, with `poetry-2.2.1 add`), will only affect this project—not your global Python installation.

* Step 6: Open the Jupyter Notebooks included in the project.
    * `jupyter lab`
* Step 7: Run the python files included in the code editor.
    * `python file_name.py`
    * NOTE: Remember that you have to create the .env file with your confidential credentials. You can use the .env.example file as reference.
* Step 8: Exiting the Poetry Shell.
    * `exit` 
* Step 9: Re-activating Poetry Shell.
    * `poetry-2.2.1 env activate`
